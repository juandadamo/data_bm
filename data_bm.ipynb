```python
import wbgapi as wb   # I use wb as a namespace in all my work
import pandas as pd   # optional but highly recommended for wbgapi
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib widget
```

```python
años = range(1950,2023)

data_gini = wb.data.DataFrame('SI.POV.GINI', [ 'ARG'],años, columns='time',numericTimeKeys=True)
data_pbi_pc = wb.data.DataFrame('NY.GDP.PCAP.CD', [ 'ARG'],años, columns='time',numericTimeKeys=True)
años_data = data_gini.count().index
gini_x = data_gini.to_numpy().T.flatten()
pbi_x = data_pbi_pc.to_numpy().T.flatten()
```

```python
fig,ax = plt.subplots(1,2,figsize=(8,4))
fig1,ax1 = plt.subplots(1,2,figsize=(8,4))
ax2,ax3 = ax
ax1a,ax1b = ax1
ax1a.plot(pbi_x,gini_x,'-o');
ax1b.plot(pbi_x,gini_x,'-o');
for i,año in enumerate(años_data):
    if np.isnan(gini_x[i])!=True:
        ax1a.text(pbi_x[i],gini_x[i],año);
        if año > 2007:
            ax1b.text(pbi_x[i],gini_x[i],año);
ax2.plot(años_data,gini_x,'-o');
ax3.plot(años_data,pbi_x,'-o');
ax2.set_xlabel('año')
ax3.set_xlabel('año')
ax2.set_ylabel('Gini')
ax3.set_ylabel('PBI / hab')
ax1a.set_xlabel('PBI / hab')
ax1b.set_xlabel('PBI / hab')
ax1a.set_ylabel('Gini')
ax1b.set_ylabel('Gini')
ax1b.set_xlim([0.8e4,1.6e4])
ax1b.set_ylim([40,46])
for axi in ax:axi.grid()
for axi in ax1:axi.grid()
fig.tight_layout()
fig1.tight_layout()
```
